Create a list named cart with initial items: Milk, Bread, Eggs, Apple

```python
# Case Study – Shopping Cart System

# Creating shopping cart list
cart = ["Milk", "Bread", "Eggs", "Apple"]

print("Initial Cart:", cart)

# extend() - Add multiple items
cart.extend(["Banana", "Butter"])
print("\nAfter extend:", cart)

# insert() - Insert item at index 2
cart.insert(2, "Cheese")
print("\nAfter insert:", cart)

# remove() - Remove specific item
cart.remove("Apple")
print("\nAfter remove:", cart)

# pop() - Remove last item
cart.pop()
print("\nAfter pop:", cart)

# index() - Find position of Egg
print("\nIndex of Eggs:", cart.index("Eggs"))

# count() - Count occurrences of Milk
print("Count of Milk:", cart.count("Milk"))

# indexing
print("\nFirst item:", cart[0])
print("Last item:", cart[-1])

# slicing
print("\nFirst 3 items:", cart[:3])

# length
print("\nTotal items in cart:", len(cart))
```

Case Study – Password Verification System with Limited Attempts
Instructions:

Create variables: • password
• attempts
• max_attempts

Use a while loop to allow the user to enter the password.

Check whether the entered password matches the original password.

If the password is correct: • Display "Access granted!"
• Stop the loop using break

If the password is incorrect: • Increase attempt count
• Display remaining attempts

If maximum attempts are reached: • Display "Maximum attempts reached. Access denied!"

```python
# Case Study – Password Verification System with Limited Attempts

# Original password
password = "python123"

# Attempt variables
attempts = 0
max_attempts = 3

# while loop
while attempts < max_attempts:

    # Taking input
    user_input = input("Enter password: ")

    # Checking password
    if user_input == password:
        print("Access granted!")
        break

    else:
        attempts += 1
        print(f"Incorrect password, try again. ({max_attempts - attempts} attempts left)")

# Maximum attempts condition
if attempts == max_attempts:
    print("Maximum attempts reached. Access denied!")
```

Case Study – Employee Attendance Count System using continue
Instructions:

Create a list named attendance with values: Present, Absent, Present, Leave, Present

Create count variables:

• present
• absent
• leave

Use a for loop to iterate through the attendance list.

Use continue statement to: • Skip remaining conditions after counting Present employees

Count: • Total Present employees
• Total Absent employees
• Total Leave employees

Display all counts using print()

```python
# Case Study – Employee Attendance Count System using continue

# Attendance list
attendance = ["Present", "Absent", "Present", "Leave", "Present"]

# Count variables
present = absent = leave = 0


# for loop
for status in attendance:

    # Count present employees
    if status == "Present":
        present += 1
        continue

    # Count absent employees
    if status == "Absent":
        absent += 1

    # Count leave employees
    elif status == "Leave":
        leave += 1

# Display counts
print("Total Present:", present)
print("Total Absent:", absent)
print("Total Leave:", leave)
```

Case Study – Student Performance Management System
Instructions

Create two lists: students marks

Use range() to: Access and print student names and marks using index positions

Use enumerate() to: Display student index number and student name

Use zip() to: Combine student names and marks together

Display all outputs using for loops

Use proper headings with print() statements

Print the final result in a clean format

```python
# Student Marks System

students = ["John", "Sam", "David"]
marks = [90, 85, 95]

print("----- Using range() -----")

for i in range(len(students)):
    print(students[i], "-", marks[i])

print("----- Using enumerate() -----")

for index, name in enumerate(students):
    print(index, "-", name)

print("----- Using zip() -----")

for name, score in zip(students, marks):
    print(name, "-", score)
```

Case Study – Employee Salary Analysis
Instructions

Create a list called salary Store employee salaries inside the list

Use Built-in Functions to:

Find maximum salary

Find minimum salary

Find total salary

Find number of employees using len()

Use lambda with map() to: Increase each salary by 10%

Use lambda with filter() to: Display salaries greater than 50000

Display:

Original salary list

Updated salary list

High salary list

Maximum salary

Minimum salary

Total salary

```python
# Employee Salary Analysis

# Create salary list
salary = [25000, 45000, 55000, 70000, 48000, 90000]

# Display original salary list
print("Original Salary List:", salary)

# Built-in Functions
max_salary = max(salary)
min_salary = min(salary)
total_salary = sum(salary)
employee_count = len(salary)

# Increase each salary by 10% using lambda and map()
updated_salary = list(map(lambda x: x + (x * 0.10), salary))

# Filter salaries greater than 50000 using lambda and filter()
high_salary = list(filter(lambda x: x > 50000, updated_salary))

# Display Results
print("\nUpdated Salary List:", updated_salary)
print("High Salary List (>50000):", high_salary)

print("\nMaximum Salary:", max_salary)
print("Minimum Salary:", min_salary)
print("Total Salary:", total_salary)
print("Number of Employees:", employee_count)
```

Case Study – Student Marks Analysis using Function and Loop
Instructions

Create a list called marks

Create a function called marks_analysis() and pass the marks list to the function

Inside the function:

Use a for loop to display all marks

Find total marks

Find average marks

Return total and average marks, call the function, and display: Student Marks, Total Marks, Average Marks

```python
# Marks List
marks = [80, 75, 90, 85, 70]

# Function Creation
def marks_analysis(data):

    print("Student Marks :")

    # for loop
    for i in data:
       print(i)

    # Operations
    total = sum(data)
    average = total / len(data)

    # Return Values
    return total, average


# Function Call
total_marks, average_marks = marks_analysis(marks)

# Output
print("\nTotal Marks   :", total_marks)
print("Average Marks :", average_marks)
```
